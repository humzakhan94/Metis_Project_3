```python
from sqlalchemy import create_engine
import pandas as pd


engine = create_engine('postgresql://humzakhan:localhost@localhost:5432/bank')

bank_churn2 = pd.read_csv('/Users/humzakhan/Metis/Metis_Projects/Metis_Project_3/Data/churn.csv')

bank_churn2.to_sql('bank_churn2', engine, index=False)
```

```python
query = 'SELECT * FROM bank_churn2;'
df = pd.read_sql(query, engine)

df.head()
```

```python
query1 = """
SELECT * 
FROM bank_churn2 WHERE geography= 'France' AND gender= 'Female';""" 
    
df1 = pd.read_sql(query1, engine)

df1.head()
```

```python
query2 = """SELECT COUNT(CustomerID), geography FROM bank_churn2 GROUP BY geography HAVING COUNT(CustomerID) > 5"""

df2 = pd.read_sql(query2, engine)

df2.head()
```
